```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df=pd.read_csv('project_analytics.csv', encoding='unicode_escape')
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True )
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype(int)
```

```python
df.info()
```

```python
df.columns
```

```python
df.rename(columns={'Matrial_Status':'Shaadi'})
```

```python
df[['Age','Amount']].describe()
```

EDA(Exploratory Data Analysis)

```python
ax=sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False ) ['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sale=df.groupby(['Gender'], as_index=False ) ['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sale)
```

From the above graph we can conclude that Female have spent more amount than Male.

```python
df.columns
```

```python
ax=sns.countplot(x='Age Group', hue='Gender', data=df)
```

```python
ax=sns.countplot(x='Age Group', hue='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales=df.groupby(['Age Group'], as_index=False ) ['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales)
```

Most of the customers are from age group 26-35 and they have also spent the maximum amount.

```python
sales_state=df.groupby(['State'], as_index=False ) ['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,6)})
sns.barplot(x='State', y='Orders', data=sales_state)
```

```python
sales_state=df.groupby(['State'], as_index=False ) ['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,6)})
sns.barplot(x='State', y='Amount', data=sales_state)
```

```python
df.columns
```

```python
ax=sns.countplot(x='Marital_Status', data=df)
for bars in ax.containers:
    ax.bar_label(bars)  
    sns.set(rc={'figure.figsize':(2,3)})
```

```python
sales_marital=df.groupby(['Marital_Status' , 'Gender'], as_index=False ) ['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(7,6)})
sns.barplot(x='Marital_Status', y='Amount', data=sales_marital,hue='Gender')
```

From above graphs we can see that most of the buyers are married women.

```python
sales_occupation=df.groupby(['Occupation'], as_index=False ) ['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,4)})     
sns.barplot(x='Occupation', y='Amount', data=sales_occupation)
```

From above graphs we can see that most of the buyers are working in IT , Healthcare and Aviation sector.

```python
ax=sns.countplot(x='Product_Category', data=df)
sns.set(rc={'figure.figsize':(30,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product=df.groupby(['Product_Category'], as_index=False ) ['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,4)})   
sns.barplot(x='Product_Category', y='Amount', data=sales_product)   
```

From above graphs we can see that most of the sold products are from Food , Clothing and Electronic

```python
sales_product=df.groupby(['Product_ID'], as_index=False ) ['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,4)})
sns.barplot(data=sales_product, x='Product_ID', y='Orders')   
```

Conclusion :

Married women are aged group 26-35 yrs from UP , Maharastra and Karnataka working from in IT , Health and Aviation are more like buy products from Food , Clothing and Electronics category.
